## Exercise

Find the highest salary earned for 2014

```python
%%SQL
    
SELECT yearID, Salary
FROM Salaries
WHERE yearID = "2014"
ORDER BY Salary DESC
```

## Exercise

Highest salary earned for 2014 using aggregate.

```python
%%SQL
    
SELECT yearID, MAX(Salary)
FROM Salaries
GROUP BY yearID
ORDER BY yearID DESC
```

## Exercise

Number of players that batted at least once

```python
%%SQL
    
SELECT COUNT(AB)
FROM Batting
WHERE AB >=1
    
```

## Exercise

Average Salary for each team.

```python
%%SQL

SELECT teamID, AVG(Salary)
FROM Salaries
GROUP BY teamID
```

## Exercise

Names of all players that have hit at least 50 home runs (HR) in a season.

```python
%%SQL

SELECT People.nameFirst, People.nameLast, Batting.HR, Batting.yearID
FROM Batting JOIN People ON Batting.playerID == People.playerID
WHERE HR >= 50
```

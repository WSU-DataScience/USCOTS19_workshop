## Exercise

Find the highest salary earned for 2014

```python
%%SQL
    SELECT yearID, Salary
    FROM Salaries
    WHERE yearID = "2014"
    ORDER BY Salary DESC
```

## Exercise

Highest salary earned for 2014 using aggregate.

```python
%%SQL
    SELECT yearID, MAX(Salary)
    FROM Salaries
    GROUP BY yearID
    ORDER BY yearID DESC
```
